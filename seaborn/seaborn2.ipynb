```python
import seaborn as sb
import pandas as pd
import numpy as np
%config InlineBackend.figure_format = 'retina'
```

```python
flight = sb.load_dataset('flights')
tips = sb.load_dataset('tips')
```

# Heatmap

```python
# To use heatmap, we need to turn our df to matrix form using corr() method

tips_corr = tips.corr()

sb.heatmap(tips_corr)
```

```python
# annot argument in heatmap shows the value for each cell

sb.heatmap(tips_corr, annot=True)
```

```python
# the cmap argument chooses color

sb.heatmap(tips_corr, annot=True, cmap='coolwarm')
```
