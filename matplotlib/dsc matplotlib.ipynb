```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,10,20)
y = x**2
plt.plot(x,y)
plt.title('my first plot')
plt.xlabel('x-axis')
plt.ylabel('y-label')
```

```python
# Creating subplot 
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(y,x, 'r')
```

```python
# 2 Object oriented interface
fig = plt.figure()
ax = fig.add_axes([0.2,0.3,0.9,0.8])

ax.plot(x,y, 'indigo')
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-axis')
ax.set_title('The Title')
```

```python
# adding 2 figures on one canvas
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.6,0.4,0.3])

axes1.plot(x,y)
axes1.set_xlabel('X-axis for axes1')
axes1.set_ylabel('Y-axis for axes1')
axes1.set_title('title for axes1')

axes2.plot(y,x)
axes2.set_xlabel('X-axis for axes2')
axes2.set_ylabel('Y-axis for axes2')
axes2.set_title('title for axes2')
```

```python
# Empty canvas of 3 x 3 subplots
fig,ax = plt.subplots(3,3)
plt.tight_layout()

# plotting on the subplots
ax[1,1].plot(y,x, 'blue')
ax[1,1].set_title('This title')
ax[0,2].plot(y//2,x, 'red')
ax[2,0].plot(y,x, 'yellow')
```

```python
# Aspect ratio, figure size, and DPI
fig = plt.figure(figsize=(8,2),dpi=100)

ax = fig.add_axes([0.1,0.1,0.8,0.9])
ax.plot(x,y)
```

# How to save my figure

```python
fig.savefig('my_figure.png')

# Creating a legend
fig = plt.figure(figsize=(8,6), dpi=80)
ax = fig.add_axes([0.1,0.2,0.8,0.6])

ax.plot(x, x**2, label = 'plot one')
ax.plot(x,x**3, label = 'plot 2')

ax.legend()
```

# Plot Appearance

```python
fig2 = plt.figure(figsize=(8,6), dpi=60)
ax = fig2.add_axes([0.1,0.1,1,1])

ax.plot(x,y**2, 'red', linestyle = '--', linewidth = 2, marker = 'p', markersize = 8)
```
