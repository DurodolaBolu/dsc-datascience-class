```python
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%config InlineBackend.figure_format = 'retina'
```

```python
# Loading data set in seaborn
tip_df = sb.load_dataset('tips')
tip_df.head()
```

```python
sb.distplot(tip_df['total_bill'],color='red')
```

```python
# A jointplot with its various kinds
# kind 'scatter'
sb.jointplot(x='total_bill',y='tip', data=tip_df, kind='scatter')

# kind reg
sb.jointplot(x='total_bill', y='tip', data=tip_df,kind='reg')
```

```python
# pairplot
sb.pairplot(tip_df)
```

# Distribution of categorical data

```python
# Barplot
sb.barplot(x='sex', y='total_bill', data=tip_df)

#countplot - just like pandas count, checks for number of occurrences

sb.countplot(x='sex', data = tip_df)

# boxplot - used to show the distribution of categorical data

sb.boxplot(x = 'day', y='total_bill', data=tip_df)
sb.boxplot(x='day', y='total_bill', data=tip_df, hue='time')
```

```python
# Regression plot using lmplot
sb.lmplot(x='total_bill', y='tip', data=tip_df, hue='sex')
```

```python
sb.lmplot(x='total_bill', y='tip', data=tip_df, hue='sex', markers=['o','v'],scatter_kws={'s':50})
```

```python
'''hue allows us to have two diff. plots on same figure,
using columns and rows allows us to have 2 Axis on same plot, based on the 
categorical data passed'''
sb.lmplot(x='total_bill', y='tip', data=tip_df,col='sex')
```

```python
# aspect manages the zoomness of the plot
sb.lmplot(x='total_bill', y='tip', data=tip_df, col='day', row='sex', aspect=1.0)
```
